## Overlay setores trajetos

Nesse notebok é feito um overlay com os setores censitários de São Paulo e os trajetos dos ônibus com buffer aplicado.

```python
import geopandas as gpd
from utils.load_shp import load_shp
from utils.save_shp import save_shp
```

```python
gdf_trajetos = load_shp('gdf_trajetos.shp')

setores_censitarios_sp = load_shp('setores_censitarios_sp.shp')
```

```python
setores_censitarios_sp = setores_censitarios_sp.to_crs(gdf_trajetos.crs)
```

```python
gdf_overlay = gpd.overlay(setores_censitarios_sp,
                          gdf_trajetos,
                          how = 'intersection')
```

```python
save_shp(gdf_overlay, 'gdf_overlay.shp')
```

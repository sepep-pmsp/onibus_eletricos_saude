## Agregações

Nesse notebook são feitas as agregações de população afetada por ônibus, linha e distrito.

```python
from utils.load_shp import load_shp
from utils.save_csv import save_csv
```

```python
gdf_overlay_pop_afetada = load_shp('gdf_overlay_pop_afetada.shp')
```

```python
pop_afetada_onibus = (gdf_overlay_pop_afetada.groupby(['cd_onibus'], as_index=False)['pop_buffer'].sum())
```

```python
pop_afetada_linha = (gdf_overlay_pop_afetada.groupby('cd_linha', as_index=False).agg(nm_linha=('nm_linha', 'first'), nm_onibus=('nm_onibus', 'first'), pop_buffer=('pop_buffer', 'sum')))
```

```python
pop_afetada_distrito = (gdf_overlay_pop_afetada.groupby('CD_DIST', as_index = False).agg({'NM_DIST': 'first',
                                                                                          'pop_buffer': 'sum'}))
```

```python
save_csv(pop_afetada_onibus, 'pop_afetada_onibus.csv')

save_csv(pop_afetada_linha, 'pop_afetada_linha.csv')

save_csv(pop_afetada_distrito, 'pop_afetada_distrito.csv')
```

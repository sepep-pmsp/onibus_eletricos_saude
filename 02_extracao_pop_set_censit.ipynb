## Extração população setores censitários

Nesse notebook é extraída a base de dados do IBGE de área territorial efetivamente domiciliada e densidade demográfica ajustada dos setores censitários.

Conforme o arquivo "Leia-me" da base de dados, disponível abaixo, a variável v0001 é o total de pessoas.

https://geoftp.ibge.gov.br/organizacao_do_territorio/malhas_territoriais/malhas_de_setores_censitarios__divisoes_intramunicipais/censo_2022/Leia_me_Area_efetivamente_ocupada.pdf

```python
import pandas as pd
import os
import requests
from io import BytesIO
from utils.save_csv import save_csv
```

```python
file_pop_set_censit = 'pop_set_censit.csv'

url_pop_set_censit = 'https://geoftp.ibge.gov.br/organizacao_do_territorio/malhas_territoriais/malhas_de_setores_censitarios__divisoes_intramunicipais/censo_2022/Area_efetivamente_domiciliada_e_densidade_ajustada_dos_Setores_Censitarios.xlsx'
```

```python
if os.path.exists(file_pop_set_censit):

    pop_set_censit = pd.read_csv(file_pop_set_censit,
                                 sep = ';',
                                 encoding='latin1')

else:

    response = requests.get(url_pop_set_censit)

    pop_set_censit = pd.read_excel(BytesIO(response.content))

    save_csv(pop_set_censit, file_pop_set_censit)
```

## Tratamento setores censitários

Nesse notebook é tratada a malha do IBGE de setores censitários do estado de São Paulo.

```python
from utils.load_csv import load_csv
from utils.load_shp import load_shp
from utils.save_shp import save_shp
```

```python
setores_censitarios = load_shp('setores_censitarios.shp')

pop_set_censit = load_csv('pop_set_censit.csv')
```

Filtro para o município de São Paulo.

```python
setores_censitarios_sp = setores_censitarios[setores_censitarios['CD_MUN'] == '3550308'].copy()
```

```python
setores_censitarios_sp['CD_MUN'].unique()
```

Merge para coluna de população.

```python
pop_set_censit = pop_set_censit.rename(columns={'v0001': 'pop'})
```

```python
setores_censitarios_sp['CD_SETOR'] = setores_censitarios_sp['CD_SETOR'].astype(str)

pop_set_censit['CD_SETOR'] = pop_set_censit['CD_SETOR'].astype(str)
```

```python
setores_censitarios_sp = setores_censitarios_sp.merge(pop_set_censit[['CD_SETOR', 'pop']],
                                                      on = 'CD_SETOR',
                                                      how = 'left')
```

```python
save_shp(setores_censitarios_sp, 'setores_censitarios_sp.shp')
```
